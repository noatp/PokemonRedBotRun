```python
from pyboy import PyBoy
import time

MAP_ID_ADDR   = 0xD35E
PLAYER_Y_ADDR = 0xD361
PLAYER_X_ADDR = 0xD362

def read_player_state(pyboy: PyBoy):
    map_id = pyboy.memory[MAP_ID_ADDR]
    y = pyboy.memory[PLAYER_Y_ADDR]
    x = pyboy.memory[PLAYER_X_ADDR]
    return map_id, int(x), int(y)

# Start PyBoy with a real window and keyboard input enabled
pyboy = PyBoy(
    "red.gb",
    window="SDL2",   # ðŸ‘ˆ actual GUI
    no_input=False,  # allow keyboard input
)

# Load your starting-in-house state
with open("pokemon_red_start.state", "rb") as f:
    pyboy.load_state(f)

# Let it settle a bit
for _ in range(10):
    pyboy.tick()
```

```python
for _ in range(360):
    pyboy.tick()

map_id, x, y = read_player_state(pyboy)
print(f"map_id = {map_id}, player_x = {x}, player_y = {y}")
```
