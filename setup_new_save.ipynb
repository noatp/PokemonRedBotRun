```python
from pyboy import PyBoy

pyboy = PyBoy("red.gb", window="SDL2")

for _ in range(3600):
    pyboy.tick()
    # Use keyboard in the SDL2 window
    # When you reach your desired state,
    # press the "Interrupt" button in Jupyter.
with open("pokemon_red_start.state", "wb") as f:
    pyboy.save_state(f)

pyboy.stop()
```

```python
pyboy = PyBoy("red.gb", window="null")

with open("pokemon_red_start.state", "rb") as f:
    pyboy.load_state(f)

for _ in range(60):
    pyboy.tick()

display(pyboy.screen.image)
pyboy.stop()
```
